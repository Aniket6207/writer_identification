```python
import cv2
import numpy as np
```

```python
img = cv2.imread("char_0.png")
```

```python
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
gray=np.float32(gray)
```

```python
harris=cv2.cornerHarris(gray,2,3,0.04)
```

```python
img[harris>0.01*harris.max()]=[255,0,0]
```

```python
cv2.imshow("Harris",img)
cv2.waitKey(0)
```

```python
img = cv2.imread("char_42.png",0)
d=cv2.FastFeatureDetector_create(50)
kp=d.detect(img,None)
i2=cv2.drawKeypoints(img,kp,None,flags=0)
cv2.imshow('Corners',i2)
cv2.waitKey(0)
```

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
 
img = cv.imread('char_0.png',0)
 
# Initiate ORB detector
orb = cv.ORB_create(50)
 
# find the keypoints with ORB
kp = orb.detect(img,None)
 
# compute the descriptors with ORB
kp, des = orb.compute(img, kp)
 
# draw only keypoints location,not size and orientation
img2=img.copy()
img2 = cv.drawKeypoints(img2, kp, None, flags=0)
plt.imshow(img2), plt.show()
```

```python
import numpy as np
import cv2 as cv
img1 = cv.imread('test55.jpeg')
gray1= cv.cvtColor(img1,cv.COLOR_BGR2GRAY)
sift = cv.SIFT_create(1000)
kp1, des1 = sift.detectAndCompute(gray1,None)
img1=cv.drawKeypoints(gray1,kp,img1)
cv.imwrite('sift_keypoints.jpg',img1)
```

```python
import numpy as np
import cv2 as cv
img2 = cv.imread('ideal_k.jpeg')
gray2= cv.cvtColor(img2,cv.COLOR_BGR2GRAY)
sift = cv.SIFT_create(1000)
kp2, des2 = sift.detectAndCompute(gray2,None)
img2=cv.drawKeypoints(gray2,kp,img2)
cv.imwrite('sift_keypoints2.jpg',img2)
```

```python
matcher=cv.DescriptorMatcher_create(cv.DESCRIPTOR_MATCHER_BRUTEFORCE_HAMMING)
```

```python
bf = cv.BFMatcher(cv.NORM_L1, crossCheck=True)
from matplotlib import pyplot as plt

matches = bf.match(des1,des2)
matches = sorted(matches, key = lambda x:x.distance)

img3 = cv.drawMatches(img1, kp1, img2, kp2, matches[:50], img2, flags=2)
plt.imshow(img3),plt.show()
```
