```python
import os
import shutil
import subprocess
from PIL import Image 
import cv2
```

```python
for i in os.listdir('character_matching/'):
    for j in os.listdir(f'character_matching/{i}'):
        os.remove(f'character_matching/{i}/{j}')
```

```python
for i in os.listdir('character_matching/'):
    os.rmdir(f'character_matching/{i}')
```

```python
for i in range(49):
    os.mkdir(f'character_matching/{i}')
    
```

```python
for i in range(49):
    shutil.copyfile(f'reference_font/{i}.jpg',f'character_matching/{i}/0.jpg')
```

```python
import numpy as np
import cv2 as cv
bf = cv.BFMatcher(cv.NORM_L1, crossCheck=True)
from matplotlib import pyplot as plt
img1 = cv.imread('zha.jpeg')
gray1= cv.cvtColor(img1,cv.COLOR_BGR2GRAY)
sift = cv.SIFT_create(1000)
kp1, des1 = sift.detectAndCompute(gray1,None)
img1=cv.drawKeypoints(gray1,kp1,img1)
for i in range(49):
    img2 = cv.imread(f'character_matching/{i}/0.jpg')
    gray2= cv.cvtColor(img2,cv.COLOR_BGR2GRAY)
    sift = cv.SIFT_create(1000)
    kp2, des2 = sift.detectAndCompute(gray2,None)
    img2=cv.drawKeypoints(gray2,kp2,img2)
    matches = bf.match(des1,des2)
    matches = sorted(matches, key = lambda x:x.distance)

    img3 = cv.drawMatches(img1, kp1, img2, kp2, matches[:50], img2, flags=2)
#     plt.imshow(img3),plt.show()
    percent=len(matches)/(len(kp2))
    print(i)
    print(percent)
```
